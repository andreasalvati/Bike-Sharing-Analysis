## 1.5 Features class 
In order to improve the quality of the predictions, it is possible to transform wrongly encoded numerical feature in categorical ones. In this case "hr", "weekday", "mnth", "season", "weathersit", "holiday", "workingday" are encoded as numerical feature but they are clearly categorical. 

```python
# here we coerce the variable type to categorical for these features
categoryVariableList = [
    "hr",
    "weekday",
    "mnth",
    "season",
    "weathersit",
    "holiday",
    "workingday",
    "yr",
]

for var in categoryVariableList:
    hour[var] = hour[var].astype("category")
```

## 1.6 Skewness

The aim of this section is to check the skewness in the numerical variables in order to improve the accuracy of the models that will be run later on in the analysis. First, we have a look at the numerical variables thanks to a distribution plot. It is clear that several variables are not symmetric. However, we will take action only on those variables which present a skewness value greater than 0.80. Finally, To reduce the skewness we will do the square root of the selected variable (In this case log was not effective)

```python
# checking the numerical variables
numeric = list(hour.select_dtypes(include=np.number).columns.values)
numeric
```

```python
# plotting the distribution of the numerical variables
def plot_skewness():
    fig, axes = plt.subplots(nrows=2, ncols=2)
    fig.set_size_inches(12, 10)
    sns.distplot(hour["temp"].compute(), ax=axes[0][0])
    sns.distplot(hour["windspeed"].compute(), ax=axes[0][1])
    sns.distplot(hour["hum"].compute(), ax=axes[1][0])
    sns.distplot(hour["cnt"].compute(), ax=axes[1][1])

    axes[0][0].set(title="temp distribution")
    axes[0][1].set(title="windspeed distribution")
    axes[1][0].set(title="hum distribution")
    axes[1][1].set(title="cnt distribution")
```

```python
warnings.filterwarnings("ignore")
plot_skewness()
```

```python
# identifing the variables that need trasformation
hour[["windspeed", "hum", "temp", "cnt", "atemp"]].compute().apply(
    lambda x: abs(skew(x)) > 0.80
)
```

```python
# reducing the skewness using sqrt
hour["cnt"] = np.sqrt(hour.cnt)
hour.head()
```

# 2. Feature Engineering
Here we switch the original columns, with some dummy ones, so to fit the format needs of our algorithms. 

```python
# for DummyEncoder to work we need to categorize the variables first
hour = hour.categorize(
    ["hr", "weekday", "mnth", "season", "weathersit", "holiday", "workingday", "yr"]
)
```

```python
# creating the Dummies with DummyEncoder
de = DummyEncoder(
    ["hr", "weekday", "mnth", "season", "weathersit", "holiday", "workingday", "yr"]
)
hour = de.fit_transform(hour)
hour
```

## 3.1 Training test split
here we split the dataset into training and test

```python
# here we create the training set from the complete dataset we have, using the date as filtering item
X_train = hour.loc[hour["dteday"] < "2012-10-1"]
del X_train["cnt"]
del X_train["dteday"]
X_train.values.shape
```

```python
# here we create the test set from the complete dataset we have, using the date as filtering item
X_test = hour.loc[hour["dteday"] >= "2012-10-1"]
del X_test["cnt"]
del X_test["dteday"]
X_test.values.shape
```

```python
y_train = hour.loc[hour["dteday"] < "2012-10-1"]["cnt"]
y_train.values.shape
```

```python
y_test = hour.loc[hour["dteday"] >= "2012-10-1"]["cnt"]
y_test.values.shape
```

## 3.2 Regression Model
Here we run our regression model, which is the first model we want to try as if with such a simple one we get good results, we would prefer it over more complex ones. 

First we need to create a Client

```python
client = Client()
client
```

```python
lr = LinearRegression()
lr.fit(X_train.values, y_train.values)
y_pred_lr = lr.predict(X_test.values)
```

```python
mse(y_test.values, y_pred_lr)
```

```python
r2_score(y_test.values.compute(), y_pred_lr.compute())
```

```python
plt.plot(
    y_test.compute(),
    y_test.compute(),
    "r--",
    y_test.compute(),
    y_pred_lr.compute(),
    "b,",
)
plt.show()
```

## 3.3 XG Boost Model

```python
xgb = XGBRegressor()
xgb.fit(X_train.values, y_train.values)
y_pred_xgb = xgb.predict(X_test.values)
```

```python
mse(y_test.values, y_pred_xgb)
```

```python
r2_score(y_test.values.compute(), y_pred_xgb.compute())
```

```python
plt.plot(
    y_test.compute(),
    y_test.compute(),
    "r--",
    y_test.compute(),
    y_pred_xgb.compute(),
    "b,",
)
plt.show()
```

## 3.4 dask and sklearn

Here we run a Random forest model from sklearn with parallel_backend("dask") in order to improve the performance

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.externals.joblib import parallel_backend
from dask.distributed import Client
```

```python
# Here we create another client to be sure not to get runtime errors
client = Client()
client
```

```python
with parallel_backend("dask"):
    regressor = RandomForestRegressor()
    parameters = [{"n_estimators": [150, 300], "max_features": ["auto", "log2"]}]
    grid_search = GridSearchCV(estimator=regressor, param_grid=parameters)
    grid_search = grid_search.fit(X_train.compute(), y_train.values.compute())
    best_parameters_rf = grid_search.best_params_
```

```python
print(best_parameters_rf)
```

```python
with parallel_backend("dask"):
    regressor = RandomForestRegressor(300, max_features="auto")
    regressor.fit(X_train.compute(), y_train.values.compute())

    # Predicting the values
    y_pred_rf = regressor.predict(X_test.compute())
```

```python
r2_score(y_test.values.compute(), y_pred_rf)
```

```python
plt.plot(y_test.compute(), y_test.compute(), "r--", y_test.compute(), y_pred_rf, "b,")
plt.show()
```

### 3.3.1 XG Boost Model with CV using sklearn

Here we use parallel backend with dask and sklearn to optimize the hyperparamiters of the xgb model

```python
import xgboost as xgb
```

```python
with parallel_backend("dask"):
    # tuning Hyperparameters
    parameters_xgb = [{"n_estimators": [500, 1000], "max_depth": [4, 8]}]
    grid_search = GridSearchCV(estimator=xgb.XGBRegressor(), param_grid=parameters_xgb)
    grid_search = grid_search.fit(X_train.compute(), y_train.values.compute())
    best_parameters_xgb = grid_search.best_params_
```

```python
print(best_parameters_xgb)
```

```python
xgb = XGBRegressor(max_depth=4, n_estimators=1000, colsample_bytree=0.4)
xgb.fit(X_train.values, y_train.values)
y_pred_xgb = xgb.predict(X_test.values)
```

```python
mse(y_test.values, y_pred_xgb)
```

```python
r2_score(y_test.values.compute(), y_pred_xgb.compute())
```

```python
plt.plot(
    y_test.compute(),
    y_test.compute(),
    "r--",
    y_test.compute(),
    y_pred_xgb.compute(),
    "b,",
)
plt.show()
```
